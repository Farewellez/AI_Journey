# Gambaran Awal: Vektor & Dot Product dalam Machine Learning 

## Rumus Dasar:
# $$a \cdot b = \sum_{i=1}^{n} a_i b_i$$

## 1. Eksperimen Dasar: Operasi Vektor dengan numpy

```python
import numpy as np
import random
```

```python
# Membuat Array Vektor
arr1 = np.array([random.randint(-20,20) for _ in range(2)])
arr2 = np.array([random.randint(-20,20) for _ in range(2)])

# Membuat operasi dasar vektor
print("Membuat operasi dasar vektor:")
print(f"{arr1} + {arr2} = {arr1 + arr2}")
print(f"{arr1} - {arr2} = {arr1 - arr2}")

# Membuat perkalian skalar dalam vektor
print("Membuat perkalian skalar dalam vektor:")
print(f"2.{arr1} = {2 * arr1}")

# Dot Product
print(f"\nDot product dari {arr1} dengan {arr2} = {np.dot(arr1,arr2)}")
```

## 2. Eksperimen Visualisasi: Gambar Vektor di 2D

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
# Buat random vector
v1 = np.array([random.uniform(-10,15) for _ in range(2)])
v2 = np.array([random.uniform(-10,15) for _ in range(2)])

# Buat fungis untuk visualisasi vektor
def plot_vector(arr,color=['yellow','red']):
    plt.axvline(x=0,color='black',lw=1.5,linestyle='-')
    plt.axhline(y=0,color='black',lw=1.5,linestyle='-')

    # Menggambar vektor
    for index_color, value in enumerate(arr):
        plt.quiver(0,0,value[0],value[1],color=color[index_color],linestyle='-',angles='xy',scale_units='xy', scale=1)
    plt.xlim(-15,16)
    plt.ylim(-15,16)
    plt.grid()
    plt.show()
    

# Input parameter
plot_vector([v1,v2])
print("Vektor 1:",v1)
print("Vektor 1:",v2)
```

## 3. Eksperimen Cosine Similarity (Kemiripan Dua Vektor)

### Rumus Dasar:

## $$ \cos(\theta) = \frac{\mathbf{a} \cdot \mathbf{b}}{\|\mathbf{a}\| \|\mathbf{b}\|} $$

```python
import numpy as np
from numpy.linalg import norm
import random
```

```python
# Cosine Similarity Function
def cosine_similarity(a,b):
    result = np.dot(a,b) / (norm(a) * norm(b))
    return result

# Input variabel
v1 = np.array([random.uniform(-10,15) for _ in range(2)])
v2 = np.array([random.uniform(-10,15) for _ in range(2)])
print(f"v1 = {v1}")
print(f"v2 = {v2}")
print("Cosine Similarity:",cosine_similarity(v1,v2))
```

## Eksperimen dot product dalam Perceptron dari neuron sederhana

```python
import numpy as np
import random
```

```python
# Perceptron function
def perceptron(x,w,bias):
    return np.dot(x,w) + bias

# variable input, bobot dan bias
x = np.array([random.uniform(1,5) for _ in range(2)])
w = np.array([random.uniform(1,5) for _ in range(2)])
bias = 1

# start program 
print(f"Input: {x}\nBobot: {w}\nBias: {bias}")
print("Output perceptron: ",perceptron(x,w,bias))
```

## Dot product antar vector dengan tambahan plot sudut diantaranya

# $$ \theta = \cos^{-1} \left( \frac{\mathbf{a} \cdot \mathbf{b}}{\|\mathbf{a}\| \|\mathbf{b}\|} \right)$$

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
from numpy.linalg import norm
```

```python
# Membuat contoh vector sederhana
v1 = np.array([3,4])
v2 = np.array([4,3])
# Membuat contoh vector ortogonal
orto1 = np.array([10,0])
orto2 = np.array([0,10])
# Membuat contoh vector sejajar
sj1 = np.array([-2,-4])
sj2 = np.array([-4,-8])

# Menggambar 2 vector sederhana
fig, ax = plt.subplots(figsize=(10,10))
ax.quiver(0,0,v1[0],v1[1],angles='xy',scale_units='xy',scale=1,color='r',label='Vector Sederhana')
ax.quiver(0,0,v2[0],v2[1],angles='xy',scale_units='xy',scale=1,color='r')

# Menggambar 2 vector ortogonal
ax.quiver(0,0,orto1[0],orto1[1],angles='xy',scale_units='xy',scale=1,color='yellow',label='Vector Ortogonal')
ax.quiver(0,0,orto2[0],orto2[1],angles='xy',scale_units='xy',scale=1,color='yellow')

# Menggambar 2 vector sejajar
ax.quiver(0,0,sj1[0],sj1[1],angles='xy',scale_units='xy',scale=1,color='blue',label='vector sejajar')
ax.quiver(0,0,sj2[0],sj2[1],angles='xy',scale_units='xy',scale=1,color='blue')

# atur plot
ax.set_xlim(-10, 15)
ax.set_ylim(-10, 15)
ax.axhline(0, color='black',linewidth=0.5)
ax.axvline(0, color='black',linewidth=0.5)

# Menampilkan grid pada setiap 1 step
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(2))
plt.gca().yaxis.set_major_locator(ticker.MultipleLocator(2))

# Menghitung dot product dari setiap vector
dot_v1v2 = np.dot(v1,v2) # vector sederhana
dot_orto = np.dot(orto1,orto2) # vector ortogonal
dot_sejajar = np.dot(sj1,sj2) # vector sejajar

# Menghitung sudut radian dari masing-masing vector dan koversikan ke derajat
rad_v = np.arccos(dot_v1v2 / (norm(v1) * norm(v2))) # radiant dari vector sederhana
rad_orto = np.arccos(dot_orto / (norm(orto1) * norm(orto2))) # radiant dari vector ortogonal
rad_sj = np.arccos(dot_sejajar / (norm(sj1) * norm(sj2))) # radiant dari vector ortogonal
# Konversi ke derajat dari masing-masing sudut radiant 
deg_v = np.degrees(rad_v)
deg_orto = np.degrees(rad_orto)
deg_sj = np.degrees(rad_sj)

# Menggambar busur sudut dari vector sederhana
v_range = np.linspace(0,rad_v,100)
x_v = np.cos(v_range)
y_v = np.sin(v_range)
ax.plot(x_v, y_v, color='green',linestyle='--',label=f'vector sederhana:  = {deg_v:.2f}')

# Menggambar busur sudut dari vector ortogonal
orto_range = np.linspace(0,rad_orto,100)
x_orto = np.cos(orto_range)
y_orto = np.sin(orto_range)
ax.plot(x_orto, y_orto, color='green',linestyle='--',label=f'vector ortogonal:  = {deg_orto:.2f}')

# Menggambar busur sudut dari vector sejajar
sj_range = np.linspace(0,rad_sj,100)
x_sj = np.cos(sj_range)
y_sj = np.sin(sj_range)
ax.plot(x_sj, y_sj, color='green',linestyle='--',label=f'vector sejajar:  = {deg_sj:.2f}')

# show output
plt.legend()
plt.grid()
plt.show()
```
